```python
import matplotlib.pyplot as plt 
%matplotlib inline 
from matplotlib.pyplot import figure
```

```python
figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
w=0.1
x = ["LR", "SVM" ]
Bow = [0.28 ,  0.32]
Cng = [0.30 , 0.29 ]
emb = [0.25 , 0.21 ]
BowTf = [0.30 ,  0.32]
CngTf = [0.30 ,  0.30  ]
embTf = [ 0.29 , 0.24 ]

bar1 = [0,1]
bar2 = [i+w for i in bar1]
bar3= [i+w for i in bar2]
bar4 = [i+w for i in bar3]
bar5= [i+w for i in bar4]
bar6= [i+w for i in bar5]

plt.bar (bar1 ,Bow, w, label="BOW" )
plt.bar (bar2 ,BowTf, w, label="BOW+TF" )
plt.bar(bar3, Cng, w , label ="CNgrams")
plt.bar(bar4, CngTf, w , label ="CNgrams+TF")
plt.bar(bar5,emb, w , label ="BertEmb")
plt.bar(bar6,embTf, w , label ="BertEmb+TF")
plt.xlabel("Benchmark versus Proposed approach on dataset1")
plt.ylabel("F1-score")
plt.ylim((0,.5))
plt.xticks(bar3, x)
plt.legend(loc="upper left")
plt.show()



```
